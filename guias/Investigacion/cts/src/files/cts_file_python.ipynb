```python
# Ejemplo: Predicción en física clásica vs física cuántica
import numpy as np

# Posición y velocidad en física clásica
posicion_clasica = 10  # metros
velocidad = 5  # m/s
tiempo = 2  # segundos
posicion_final = posicion_clasica + velocidad * tiempo
print(f"Posición clásica: {posicion_final} m")

# Probabilidad en física cuántica
estado_cuantico = np.array([0.6, 0.8])  # Estado de superposición
probabilidad = np.square(np.abs(estado_cuantico))
print(f"Probabilidad de estados: {probabilidad}")
```

```python
# Ejemplo: Análisis mixto en ciencias sociales
import pandas as pd
import matplotlib.pyplot as plt

# Datos cuantitativos: encuestas de satisfacción laboral
datos = {"Satisfacción": [3, 4, 5, 4, 2, 5, 3], "Edad": [25, 30, 35, 40, 45, 50, 55]}
df = pd.DataFrame(datos)
promedio = df["Satisfacción"].mean()
print(f"Satisfacción promedio: {promedio}")

# Datos cualitativos: análisis de narrativas
narrativas = [
    "El ambiente es positivo pero falta reconocimiento.",
    "Los salarios son bajos pero el equipo es solidario.",
    "Me siento valorado, aunque podría haber más flexibilidad."
]
for i, texto in enumerate(narrativas, 1):
    print(f"Entrevista {i}: {texto}")

# Visualización de resultados
plt.bar(df["Edad"], df["Satisfacción"])
plt.title("Satisfacción laboral por edad")
plt.xlabel("Edad")
plt.ylabel("Satisfacción")
plt.show()
```

```python
# Ejemplo: Evolución del presupuesto en ciencia y tecnología
import matplotlib.pyplot as plt

# Datos simulados
anios = [1980, 1990, 2000, 2010, 2020]
presupuesto = [0.8, 0.5, 0.6, 0.9, 1.2]  # Porcentaje del PBI

# Visualización
plt.plot(anios, presupuesto, marker='o', color='teal')
plt.title("Evolución del presupuesto en ciencia y tecnología")
plt.xlabel("Año")
plt.ylabel("Presupuesto (% del PBI)")
plt.grid(True)
plt.show()
```

```python
# Ejemplo: Impacto de género frente al cambio climático
import pandas as pd

# Datos simulados
datos = {
    "Grupo": ["Mujeres", "Hombres"],
    "Afectados_por_desastres": [70, 30],  # Porcentaje
    "Acceso_a_recursos": [40, 60]  # Porcentaje
}
df = pd.DataFrame(datos)

# Análisis de desigualdades
print(df)

# Visualización
df.plot(kind="bar", x="Grupo", stacked=True, color=["pink", "blue"], legend=False, figsize=(8, 5))
plt.title("Desigualdades de género frente al cambio climático")
plt.ylabel("Porcentaje")
plt.show()
```
